## Importing libraries, models and data

```python
import numpy as np
import pandas as pd
import random

random.seed(2137)

import warnings
warnings.filterwarnings("ignore")
pd.options.mode.chained_assignment = None

import pickle
rf = pickle.load(open('RandomForest', 'rb'))
gb = pickle.load(open('GradientBoosting', 'rb'))
dt = pickle.load(open('DecisionTree', 'rb'))

X = pickle.load(open('X', 'rb'))
y = pickle.load(open('y', 'rb'))

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

## Permutational importance 

First things first I am going to check permutation importance of the best performing model - `Gradient Boosting`.

```python
import eli5
from eli5.sklearn import PermutationImportance

perm_gb = PermutationImportance(gb, random_state=1).fit(X_train, y_train)
eli5.show_weights(perm_gb, feature_names = X_train.columns.tolist())
```

We can see that the most important feature is `median income`, followed right by `longitude` and `latitude` - our coordinates. In the top 5 variables there also is `ocean proximity` - another feature connected with localisation. This means that `Gradient Boosting` model puts big emphasis on localisation of households. We can also observe that the variables we created are far more important than base variables. It is worth checking whether the model would give better scorings when trained without those base features.

## Permutational importance for other models

```python
perm_rf = PermutationImportance(rf, random_state=1).fit(X_train, y_train)
eli5.show_weights(perm_rf, feature_names = X_train.columns.tolist())
```

```python
perm_dt = PermutationImportance(dt, random_state=1).fit(X_train, y_train)
eli5.show_weights(perm_dt, feature_names = X_train.columns.tolist())
```

Permutation importance for those two models provide us with useful information. In both of them `median income` plays a big role, with other features being somehow far behind it, which is different from previous model. All the other observations made are very similar to the `gradient boosting` model. In the top 5 we can find all the variables connected with localisation (`longitude`, `latitude` and `ocean proximity`). New variables are also more important than the variables they were based on. It leads us to similar conclusion - maybe it is worth it to get rid of old variables?
